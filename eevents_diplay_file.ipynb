```python
import numpy as np
import sys
import re
import os
import cygnus_lib as cy
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import pandas as pd
import time
from IPython import display
%matplotlib inline
```

```python
folder_path = './cos/'
files_in_folder = os.listdir(folder_path)
```

```python
cy.set_atlas_style('square')
for iTr in range(0, len(files_in_folder)):
    try:
        image = cy.read_image_h5(folder_path+files_in_folder[iTr])
        #plt.title("I%d Run%d" % (iTr, run))
        plt.imshow(image, cmap="gray", vmin=85,vmax=140)
        time.sleep(0)
        plt.plot()
        display.display(plt.show())
        display.clear_output(wait=True)
    except KeyboardInterrupt:
        break
```
