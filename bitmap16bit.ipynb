```python
from PIL import Image
# il problema e' il bitmap a 16 bit che non c'e' nei formati standard di PIL 
# https://pillow.readthedocs.io/en/3.3.x/handbook/concepts.html#concept-modes
# ma fra le righe del codice si trova:
# https://pillow.readthedocs.io/en/3.0.0/_modules/PIL/Image.html

import numpy as np
import matplotlib.pyplot as plt
with open('/Volumes/Cygno4/Cygnus/Data/LAB/Data_Camera/BMP/2Cygni.bmp', 'rb') as f:
    datab = f.read()
pix = Image.frombytes("I;16", (2048, 2048), datab)
image = np.array(pix)
plt.imshow(image, cmap="gray")
plt.show()
```

```python
image = np.loadtxt('/Volumes/Cygno4/Cygnus/Data/LAB/Data_Camera/BMP/2Cygni.dat', dtype='uint16', delimiter='\t')
plt.imshow(np.array(image), cmap="gray")
plt.show()
```
